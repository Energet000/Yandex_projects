# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

Описание данных
 - airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
 - balcony — число балконов
 - ceiling_height — высота потолков (м)
 - cityCenters_nearest — расстояние до центра города (м)
 - days_exposition — сколько дней было размещено объявление (от публикации до снятия)
 - first_day_exposition — дата публикации
 - floor — этаж
 - floors_total — всего этажей в доме
 - is_apartment — апартаменты (булев тип)
 - kitchen_area — площадь кухни в квадратных метрах (м²)
 - last_price — цена на момент снятия с публикации
 - living_area — жилая площадь в квадратных метрах(м²)
 - locality_name — название населённого пункта
 - open_plan — свободная планировка (булев тип)
 - parks_around3000 — число парков в радиусе 3 км
 - parks_nearest — расстояние до ближайшего парка (м)
 - ponds_around3000 — число водоёмов в радиусе 3 км
 - ponds_nearest — расстояние до ближайшего водоёма (м)
 - rooms — число комнат
 - studio — квартира-студия (булев тип)
 - total_area — площадь квартиры в квадратных метрах (м²)
 - total_images — число фотографий квартиры в объявлении

## Предобработка данных

### Обработка аномальных значений

**Столбец *rooms***

Присутсвуют нулевые значения, которые не отражают действительности. Возможно, ошибки вызваны тем, что пользователи забыли указать число комнат и использовалось значение по умолчанию или ошиблись при заполнениие значения. В данных присутсвуют редкие, но чрезмерно большие значения, поэтому воспользуемся медианой.

```python
# Заменим нули на медиану
df.loc[df['rooms'] == 0, 'rooms'] = df['rooms'].median()
```

```python
# Проверим возможные значения
df['rooms'].unique()
```

**Столбец *ceiling_height***

Присутсвуют аномально высокие значения (больше 10 м), которые могли появится из-за мисклика при постановке запятой, разделяющей разряды. Для решения проблемы можно значения выбросов поделить на 10.

```python
# Для значений в диапазаонеот 10 м до 99 м
df.loc[df['ceiling_height'] >= 10, 'ceiling_height'] = df.loc[df['ceiling_height'] >= 10, 'ceiling_height'] / 10
# Для значений >= 100 м
df.loc[df['ceiling_height'] >= 10, 'ceiling_height'] = df.loc[df['ceiling_height'] >= 10, 'ceiling_height'] / 10
# Проверим значения 
df['ceiling_height'].hist()
```

### Обработка пропусков

```python
# Подсчитаем количество пропусков
print(df.isnull().sum())
```

```python
# Подсчитаем долю пропусков
print(df.isnull().mean())
```

**Столбец *ceiling_height***

Пропуски могли возникнуть из-за нежелания измерить высоту потолка и записать полученное значение. Можно заменить пропуски самими повторяющимися значениями. Наиболее частое значение найдем с помощью методов *value_counts* и *index*.

```python
# Заполним пропуски
df['ceiling_height'].fillna(df['ceiling_height'].value_counts().index[0], inplace=True)
```

Люди могли не знать точное число этажей и поленились посмотреть на кнопки в лифте. Предлагаю заменить пропуски самым встречаемым значением.

**Столбец *is_apartment***

Пропуски могли возникнуть из-за того, что люди не знали чем апартаменты отличаются от обычного жилья. Доля пропусков составляет 88%, поэтому имеет смысл удалить этот столбец целиком.


```python
df.drop(axis=1, columns='is_apartment', inplace=True)
```

**Столбец *living_area***

Возможно, некоторые пользователи не знали точную жилую площадь или посчитали данную информацию незначительной.

```python
# Посмотрим как жилая площадь коррелирует с другими параметрами
df.corr().loc['living_area']
```

Жилая площадь ожидаемо сильно коррелирует с количеством комнат, поэтому заполним пропуски средним значением, сгруппировав данные по количеству комнат.

```python
# Напишем цикл, где для каждого значения количества комнат заполним пропуски средним значением жилой площади
for room in df['rooms'].unique():
    df.loc[df['rooms'] == room, 'living_area'] = df.loc[df['rooms'] == room, 'living_area'].fillna(
        df.loc[df['rooms'] == room, 'living_area'].mean()
    )     
```

**Столбец *kitchen_area***

Возможно, некоторые пользователи не знали точную площадь кухни или посчитали данную информацию незначительной.

```python
# Посмотрим как площадь кухни коррелирует с другими параметрами
df.corr().loc['kitchen_area']
```

Площадь кухни сильно коррелирует с общей площадью. Посмотрим во сколько раз кухня меньше все квартиры.

```python
# Посмотрим во сколько раз общая площадь квартиры больше площади кухни
(df['total_area'] / df['kitchen_area']).hist()
# Найдем моду распределения
mode = (df['total_area'] / df['kitchen_area']).mode()[0]
print(f'Мода распределения: {mode}')
```

Таким образом можно считать, что площадь кухни зачастую в 5 раз меньше общей площади. На основе этого заполним пропуски.

```python
df['kitchen_area'].fillna(df['total_area'] / mode, inplace=True)
```

## Расчёты и добавление результатов в таблицу

**Этаж квартиры**

Реализуем функцию, которая принимает на вход номер этажа и общее число этажей, а на выходе возвращает одно из значений (первый, последний, другой)

```python
# На вход функции подется строка из таблицы
def floor_name(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 'первый'
    if floor == floors_total:
        return 'последний'
    return 'другой'

# Применим функцию к таблице
df['floor_name'] = df.apply(floor_name, axis=1)
```

**Cоотношение жилой и общей площади, а также отношение площади кухни к общей**

```python
# Создадим необходимые столбцы
df['living_to_total'] = df['living_area'] / df['total_area']
df['kitchen_to_total'] = df['kitchen_area'] / df['total_area']
```
